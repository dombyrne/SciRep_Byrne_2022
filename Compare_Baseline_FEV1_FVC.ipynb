```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
import statsmodels.api as sm
from scipy.stats import mannwhitneyu
import matplotlib.pyplot as plt
```

```python
#load main table
df = pd.read_csv('../results/copd_table_post_filtering.tsv', sep='\t', index_col='eid')

#load spiro data
fev_fvc_f = 'ukb_fev_fvc.tsv'
fev_fvc = pd.read_csv(fev_fvc_f, sep='\t', index_col='eid')
fev_fvc['FEV1_FVC'] = fev_fvc['FEV1_best'] / fev_fvc['FVC_best']
fev_fvc = fev_fvc[fev_fvc['spiro_reproducible'] == 1].dropna()

#join tables
df = df.join((fev_fvc, height, cardiac_op))
df.dropna(subset=['FEV1_FVC'], inplace=True)
df
```

```python
df['ethnicity'].value_counts()
```

```python
#compare fev1/fvc at recruitment - fully adjusted
form1 = 'FEV1_FVC ~ chd + sex + age_at_ac1'
mod1 = smf.ols(data=df, formula=form1)
res1 = mod1.fit()
print(res1.summary())
print(res1.pvalues)
```

```python
#compare fev1/fvc at recruitment - fully adjusted
form2 = 'FEV1_FVC ~ chd + sex + age_at_ac1 + Townsend_DI + bmi + smoking_status_1 + smoking_status_2'
mod2 = smf.ols(data=df, formula=form2)
res2 = mod2.fit()
print(res2.summary())
print(res2.pvalues)
```

```python
#split into two, based on those with and without diagnoses of asthma/copd pre-AC
cols = ['FEV1_FVC', 'FVC_best', 'FVC_pc_pred', 'FEV1_pc_pred', 'chd', 'sex', 'age_at_ac1', 'height', 'cardiac_op_pre_ac', 'Townsend_DI', 'bmi', 'smoking_status_1', 'smoking_status_2']
ndf = df[cols].copy()

ukb_gen_f = 'ukb_general_info.tsv'
ac_date = pd.read_csv(ukb_gen_f, sep='\t', usecols=['eid', 'ac1_date'], index_col='eid')
ac_date['ac1_date'] = ac_date['ac1_date'].astype('datetime64')

copd_date_f = 'copd_event_dates.tsv'
copd_date = pd.read_csv(copd_date_f, sep='\t', usecols=['eid', 'event_date'], index_col='eid')
copd_date = copd_date.rename(columns={'event_date': 'copd_date'})
copd_date['copd_date'] = copd_date['copd_date'].astype('datetime64')

asthma_date_f = 'asthma_dates.tsv'
asthma_date = pd.read_csv(asthma_date_f, sep='\t', usecols=['eid', 'asthma_date'], index_col='eid')
asthma_date['asthma_date'] = asthma_date['asthma_date'].astype('datetime64')

ndf = ndf.join([ac_date, copd_date, asthma_date])

no_obs_filt = ~(ndf['copd_date'] < ndf['ac1_date']) & ~(ndf['asthma_date'] < ndf['ac1_date'])

ndf_no_obs = ndf[no_obs_filt]
```

```python
#compare fev1/fvc in participants without asthma/copd diagnoses prior to the assessment centre - age/sex adjusted
form1 = 'FEV1_FVC ~ chd + sex + age_at_ac1'
mod1 = smf.ols(data=ndf_no_obs, formula=form1)
res1 = mod1.fit()
print(res1.summary())
print(res1.pvalues)
```

```python
#compare fev1/fvc in participants without asthma/copd diagnoses prior to the assessment centre - fully adjusted
form2 = 'FEV1_FVC ~ chd + sex + age_at_ac1 + Townsend_DI + bmi + smoking_status_1 + smoking_status_2'
mod2 = smf.ols(data=ndf_no_obs, formula=form2)
res2 = mod2.fit()
print(res2.summary())
print(res2.pvalues)
```
